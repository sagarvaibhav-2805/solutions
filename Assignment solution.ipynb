```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SBIN_Data-2.zip')
```

```python
l = [a for a in df["Date"].unique()]
l1 = [a for a in df["Time"].unique()]
```

```python
rank = []
order = []

for i in l:
    for k in l1:
        volume = df[(df["Date"] == i) & (df['Time'] == k)]['Volume'].values[0]
        rank=[]
        for j in range(l.index(i), l.index(i)-6, -1):
            if j < 0:
                break
            rank.extend(df[(df["Date"] == l[j]) & (df['Time'] == k)]['Volume'].values)
        rank.sort(reverse=True)
        a = rank.index(volume)+1
        order.append(a)
```

```python
df['rank'] = order
```
